```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
pen = sns.load_dataset('penguins')

x = np.linspace(-np.pi, np.pi, 50)
y1 = np.cos(x)
y2 = np.sin(x)
```

```python
# functional

plt.plot(x,y1, label = 'cosine')
plt.plot(x, y2, label = 'sine')
plt.xlabel('values of x')
plt.ylabel('f(x)')
plt.title('Plot of Sine and Cosine Between -Pi and Pi')
plt.legend(loc = (1.1,0))
# to have graph appear in a notebook environment, use
plt.show
```

```python
# object oriented

# plt.subplot() automatically instantiates these objects for us
# they are from a class in matplotlib

fig, ax = plt.subplots(figsize = (3,3))

ax.plot(x, y1, label = 'cosine')
ax.plot(x, y2, label = 'sine')
ax.set_xlabel('x value')
ax.set_ylabel('f(x)')
ax.set_title('Plot of Sine and Cosine Between -pi and pi')
ax.legend(loc = (1.1,0))
```

```python
# functional
# assigning it to a variable this time to save the plot

fig = plt.figure(figsize=(4,4))
plt.scatter(data = iris,x = 'sepal_length', y = 'sepal_width', 
            c = "pink")

# used to have to do 
# plt.scatter(iris['sepal_length'], iris['sepal_width'])

fig.savefig('pink.png')
```

```python
# seaborn has better labeling defaults
# example gallery https://seaborn.pydata.org/examples/index.html

sns.scatterplot(data = iris, x ='sepal_length', 
                y ='sepal_width', hue = 'species')
plt.legend(loc = (1.1,0))
```
