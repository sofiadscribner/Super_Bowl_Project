```python
# read YouTube API key from file

with open('youtube_api_key.txt', 'r') as file:
    yt_key = file.read().strip()  # Remove extra spaces/newlines

# base YouTube API URL

url = 'https://www.googleapis.com/youtube/v3/videos'

# list of video IDs (YouTube API allows max 50 per request)

videos = ['vAf5TjiUJgo','g0D5MU6_0n0', 'DXfWY8GpX8Y', '7lm5C0rVV3o', '4fC-1xIZzXs', 
          'XzHzbKDBKkg', '1oKdZuQB5xE', 'f_EzJKJXtzM', 'F1dvX9Vs0ns', 'Fi6ZRL_rWc8', 
          'd0DVNqSRMyU', 'FHpJSdV-hsE','H3EsFtUCOno', 'Neul4wINfUQ', '87O-Q0KW1zg', 
          'RmsC7th22yw', 'y21gJsdkAd8', '8J1iVYkFh7k', 'VHjNKVwIHfs', 'uqmWNiey3kU', 
          'sZtdfuNZtIU', 'Z03aMedKz04', 'ubSnD737JaM', 'xiadPtcevQI', 'koZySHoZlH4', 
          't5425luqI5s', 'TX1GEA79fNQ', 'J4rYngvvUB4', 'kwUJq0DA0xo', 'ZTBh6wfMG6', 
          'W_xrL5vRw5c', '5_0HcGEjAzM', '5OTvykg-PyM', 'noJXHyBPvnQ', 'NBuB85VsZfI', 
          '1kETt59yn6A', 'YKzk1h7qrj8', 'fqunTCZJ8baU', 'PjCSipST46I', 'YVeCwSWNLrM', 
          'VeCwSWNLrM', 'FQcDTAlYAx4', 'wYPTZIFQoDQ', 'OBcd8S0YYuU', 'EgaRKjuYtRg', 
          'zoP_UON723Y', 'vQ88JB-lgtQ', 'RBbl-zh7pHQ', 'r-6sqrovm8U', 'M5EYq8wiC-4', 
          'ZBT_7eqT8bA','8DSWtvNMXfw', '4mPKDEfo0UM', 'p4OGQQPMiXQ', 'yDpVLa824Wg', 
          'LZvtROpJvqc','67Vbrm9IK6A', 'dSJtfg3GWFE', '7glIC7OaIro', 'D6HUxANY8z8', 
          '94BqlDQ-Ppo', 'z5fcvlWdfdA', 'nvXJJRjGVT4']

# function to chunk the list into groups of 50

def chunk_list(lst, size):
    for i in range(0, len(lst), size):
        yield lst[i:i + size]

video_data = []

# send requests in chunks of 50

for chunk in chunk_list(videos, 50):
    params = {
        'part': 'snippet, statistics',
        'id': ','.join(chunk),
        'key': yt_key
    }
    
    # make API request

    response = requests.get(url, params=params)

    # parse JSON response

    data = response.json()
    for item in data.get("items", []):
        video_id = item["id"]
        title = item["snippet"]["title"]  # Extract title
        stats = item["statistics"]
        views = stats.get("viewCount", 0)
        likes = stats.get("likeCount", 0)
        video_data.append({"Video ID": video_id, "Title": title, "Views": int(views), "Likes": int(likes)})


# convert list of dictionaries into a DataFrame

df = pd.DataFrame(video_data)

# looks great! save locally

#df.to_csv("youtube_data.csv", index=False)
```
